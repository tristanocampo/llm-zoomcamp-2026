## First Setup

```python
import os
from dotenv import load_dotenv
from google.genai import Client

# Load the .env file
load_dotenv()
print(f"Key found: {os.environ.get('GEMINI_API_KEY') is not None}")
# print(f"Key value: {os.environ.get('GEMINI_API_KEY')}")
```

```python
# Initialize the client
client = Client(api_key=os.environ.get('GEMINI_API_KEY'))
```

RAG (Retrieval-Augmented Generation) is an AI framework that improves LLM responses by first fetching relevant information from an external knowledge source (like a database or document store) and then feeding that retrieved data as context to the LLM to generate an accurate, up-to-date answer.

```python
# Test the connection
response = client.models.generate_content(
model="gemini-2.5-flash",
contents="Hello! Is my environment working?"
)
print(response.text)
```

## Fetching FAQs of DataTalks Courses

```python
import requests

import requests

docs_url = "https://datatalks.club/faq/json/courses.json"
response = requests.get(docs_url)
courses_raw = response.json()
print(courses_raw)
```

```python
documents = []
url_prefix = "https://datatalks.club/faq"

for course in courses_raw:
    course_url = f"""{url_prefix}{course["path"]}"""

    course_response = requests.get(course_url)
    course_response.raise_for_status()
    course_data = course_response.json()

    documents.extend(course_data)

len(documents)
print("Sample retrieved document:", documents[0])
```

Each entry has:

- `id` - unique identifier
- `course` - course slug (e.g., `machine-learning-zoomcamp`)
- `section` - which section of the course
- `question` - the FAQ question
- `answer` - the FAQ answer
